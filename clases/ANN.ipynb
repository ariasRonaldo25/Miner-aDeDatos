```python
# Cómo importar las librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importar el data set
dataset = pd.read_csv('/content/gdrive/My Drive/datos_DL/Churn_Modelling.csv')

X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
```

```python
# Codificar datos categóricos
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X_2 = LabelEncoder()
X.iloc[:, 2] = labelencoder_X_2.fit_transform(X.iloc[:, 2])
```

```python
X
```

```python
# crear un dataframe de variables dummies
dummy_Geography = pd.get_dummies(X["Geography"], prefix="geo",drop_first=True)
# el argumento drop_first elimina la primera variable para evitar colinealidad

# print(dummy_country)

# Borrar la variable original
X = X.drop(["Geography"], axis = 1)

# Concatenar las variables dummies con el dataframe original
X = pd.concat([dummy_Geography, X], axis = 1)

X.head()
```

```python
# Dividir el data set en conjunto de entrenamiento y conjunto de testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Escalado de variables
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
np.mean(X_train)
```

```python
np.std(X_train)
```

# Parte 2 - Construir la RNA

```python
# Importar Keras y librerías adicionales
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Inicializar la RNA
classifier = Sequential()
```

```python
# Añadir las capas de entrada y primera capa oculta
classifier.add(Dense(units = 6, kernel_initializer = "uniform",  
                     activation = "relu", input_dim = 11))
```

```python
# Añadir la segunda capa oculta
classifier.add(Dense(units = 6, kernel_initializer = "uniform",  activation = "relu"))
```

```python
# Añadir la capa de salida
classifier.add(Dense(units = 1, kernel_initializer = "uniform",  activation = "sigmoid"))
```

```python
# Compilar la RNA
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
# Ajustamos la RNA al Conjunto de Entrenamiento
classifier.fit(X_train, y_train,  batch_size = 10, epochs = 100)
```

# Parte 3 - Evaluar el modelo y calcular predicciones finales

```python
# Predicción de los resultados con el Conjunto de Testing
y_pred  = classifier.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
# Elaborar una matriz de confusión
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
