```python
# EDA TRADICIONAL.
from pandas import read_csv
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = read_csv('boston.csv', header=None, delimiter=r"\s+", names=column_names)
print(data.head(5)) 
```

```python
#Variables independientes
X=data.iloc[:,0:-1]
Y=data.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Escalado de variables
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X = sc_X.fit_transform(X)
Y = sc_Y.fit_transform(y.reshape(-1,1))
```

```python
X
```

```python
# Ajustar la regresión con el dataset
from sklearn.svm import SVR
regression = SVR(kernel = "rbf")
regression.fit(X, Y)
model=regression.fit(X_train, Y_train)
y_predict_train=model.predict(X_train)
y_predict_test=model.predict(X_test)
```

```python
# Predicción de nuestros modelos con SVR
y_predict_train=model.predict(X_train)
y_predict_test=model.predict(X_test)
```

```python
print('Y predict entrenamiento:',y_predict_train)
```

```python
print('Y predict pruebas:',y_predict_test)
```

```python
from sklearn.metrics import r2_score
print('R2 para entrenamiento:', r2_score(Y_train,y_predict_train))
print('R2 para pruebas:', r2_score(Y_test,y_predict_test))
```
